```python
#Importing related dependencies;
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading Dataset;
df = pd.read_csv('dataset/Salary Data.csv')
```

```python
#Checking dataset;
df.head(10)
```

```python
#Number of Rows and Cols in Dataset
df.shape
```

```python
#info about Dataset
df.info()
```

```python
#Checking null values
df.isnull().sum()
```

```python
#remove null values
df = df.dropna()
```

```python
#Checking again
df.isnull().sum()
```

```python
#Removing spaces in str_columns;
df.columns = df.columns.str.replace(' ', '_')
```

```python
#Educational information by gender
female_edu=df[df["Gender"]=="Female"]["Education_Level"].value_counts()
male_edu=df[df["Gender"]=="Male"]["Education_Level"].value_counts()
```

```python
#Displaying information in subgraphs
plt.subplot(1,2,1)
plt.pie(female_edu, labels=female_edu.index, autopct='%1.1f%%', explode=[0.1,0,0], shadow=True)
plt.title("Education rate of women")
plt.subplot(1,2,2)
plt.pie(male_edu, labels=male_edu.index, autopct='%1.1f%%', explode=[0.1,0,0], shadow=True)
plt.title("Education rate of men")
plt.legend(["Bachelor's", "PhD","Master's"], loc="upper right", bbox_to_anchor=(1.7, 1))
plt.figsize=(10,50)
```

```python
#Sorting by salary and selecting the first 50 people
df1 = df.sort_values(by='Salary',ascending=False).head(50)
df2 = df1['Gender'].value_counts()
#Creation of pie chart
plt.pie(df2, labels=df2.index, autopct='%1.1f%%', shadow=True)
plt.title("Gender distribution of the 50 highest paid people")
```

```python
#Examining the relationship between experience and salary by gender
sns.relplot(data=df, x=df['Years_of_Experience'],y=df['Salary'],hue='Gender')
```

```python
#Lines containing the word 'market' were filtered out.
df_job = df[df["Job_Title"].str.contains("market", case=False)]
#info about new dataset
df_job.info()
```

```python
#In order to make a more accurate assessment, people working as a junior were excluded from the list
junior_rows = df_job[df_job['Job_Title'].str.contains('Junior')]
df_new = df_job[~df_job['Job_Title'].str.contains('Junior')]
```

```python
#Checking dataset;
junior_rows.info(), df_new.info()
```

```python
df_new.head(30)
```

```python
#The relationship between education levels and salary of men and women doing similar job (Marketing)
sns.boxplot(data=df_new, x="Education_Level", y='Salary', hue="Gender")
```

```python
#Average salary of women doing similar jobs
df_new[df_new['Gender']=='Female']['Salary'].mean()
```

```python
#Average salary of men doing similar jobs
df_new[df_new['Gender']=='Male']['Salary'].mean()
```

```python
#The relationship between years of experience and salary of men and women doing similar job (Marketing)
sns.lmplot(x="Salary", y="Years_of_Experience", hue='Gender', data=df_new, height=5, aspect=1.5)
sns.set_color_codes(palette='deep')
sns.set_style("darkgrid")
```
